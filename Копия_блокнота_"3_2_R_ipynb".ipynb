```r
# Задание 7
library(ggplot2)
library(tidyr)
library(dplyr)

music_genre <- read.csv("music_genre.csv")
set.seed(31)
music_genre_sample <- music_genre[sample(nrow(music_genre), 20000, replace = FALSE),]


most_popular_genre <- names(sort(table(music_genre$music_genre), decreasing = TRUE))[1]
least_popular_genre <- names(sort(table(music_genre$music_genre), decreasing = FALSE))[1]
means_by_genre <- aggregate(music_genre[,2:14], by=list(music_genre$music_genre), mean)
sorted_genres <- sort(unique(music_genre$music_genre), decreasing = TRUE)
most_popular_genre <- sorted_genres[1]
least_popular_genre <- sorted_genres[length(sorted_genres)]
most_popular_means <- means_by_genre[means_by_genre$Group.1 == most_popular_genre, 2:14]
least_popular_means <- means_by_genre[means_by_genre$Group.1 == least_popular_genre, 2:14]

most_popular_means_long <- tidyr::gather(most_popular_means, key = "feature", value = "mean_val")
least_popular_means_long <- tidyr::gather(least_popular_means, key = "feature", value = "mean_val")

plot_data <- data.frame(
  genre = c(rep(most_popular_genre, ncol(most_popular_means)), rep(least_popular_genre, ncol(least_popular_means))),
  feature = c(names(most_popular_means), names(least_popular_means)),
  mean_val = c(most_popular_means_long$mean_val, least_popular_means_long$mean_val),
  type = rep(c("positive", "negative"), each = ncol(most_popular_means))
)

ggplot(plot_data, aes(x = feature, y = mean_val, fill = type)) +
  geom_col(position = "dodge") +
  facet_wrap(~ genre, ncol = 2) +
  scale_fill_manual(values = c("red", "grey"), labels = c("Самый популярный жанр", "Самый непопулярный жанр")) +
  theme_minimal() +
  xlab("Характеристика") +
  ylab("Среднее значение")
```

## Раздение и объединение векторов и таблиц

Разделим табличку с книгами по условию: в одной будет цена больше 100 долларов, во второй - меньше.

```r
up <- subset(books, price > 100)
down <- subset(books, price <= 100)
```

```r
nrow(up)
nrow(down)
```

Теперь объединим таблицы обратно. Для горизонтального склеивания таблиц есть функция rbind().

```r
rbind(down, up)
```

Разобьем по вертикали таблицу на две: по 4 и 2 два столбца.

```r
left <- books[1:4]
right <- books[5:6]
ncol(left)
ncol(right)
```

```r
head(left, n=2)
```

Склеим обратно с помощью cbind().

```r
head(cbind(right, left), n=2)
```

Проверим, что исходная табличка была такой же.

```r
head(books, n=2)
```

Теперь посмотрим на слияние таблиц: для этого есть функция merge().

```r
left <- books[c(1:3)]
right <- books[c(3:6)]
print(head(left, n=2))
print(head(right, n=2))
```

При слиянии итоговая табличка сортируется по возрастанию значений цены.

```r
books_merged <- merge(left, right, by="title")
head(books_merged[order(books_merged$price),], n=2)
```
